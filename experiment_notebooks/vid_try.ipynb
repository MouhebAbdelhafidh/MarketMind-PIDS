```python
import torch
from diffusers import DiffusionPipeline
import numpy as np
import imageio
from PIL import Image, ImageDraw, ImageFont
from concurrent.futures import ThreadPoolExecutor
from IPython.display import Image as IPImage, display

# Load the video diffusion model
pipe = DiffusionPipeline.from_pretrained(
    "cerspense/zeroscope_v2_576w",
    torch_dtype=torch.float16,
    use_safetensors=False  # Avoid safetensors error
).to("cuda")

# Define the prompt
prompt = "Ultra-HD promo of a makeup palette being opened in slow motion, sparkles and glowing lights, pastel themed background"
# Reduce resolution and number of frames for faster processing
height = 256  # Reduced height
width = 256   # Reduced width
num_frames = 12  # Reduced number of frames

# Generate video with reduced resolution and number of frames
output = pipe(prompt, num_frames=num_frames, height=height, width=width)

# Extract and process frames
frames = output.frames
if isinstance(frames, list):
    frames = np.array(frames)

frames = np.squeeze(frames)  # Remove unnecessary dimensions
frames = (frames * 255).clip(0, 255).astype(np.uint8)  # Convert to uint8

# Path to save the final GIF
output_path = "/content/drive/MyDrive/output_with_cta.gif"

# Get user input for CTA
cta_text = input("Enter your call-to-action (CTA) text: ")

# Load a better font and make the text much bigger
try:
    font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 72)  # Increased font size
except:
    font = ImageFont.load_default()

# Function to overlay CTA on a single frame
def add_cta_to_frame(frame):
    img = Image.fromarray(frame)
    draw = ImageDraw.Draw(img)

    # Text position (bottom center)
    bbox = draw.textbbox((0, 0), cta_text, font=font)
    text_width = bbox[2] - bbox[0]
    text_height = bbox[3] - bbox[1]
    x = (img.width - text_width) // 2
    y = img.height - text_height - 40

    # Text outline for visibility
    for dx in range(-2, 3):
        for dy in range(-2, 3):
            draw.text((x + dx, y + dy), cta_text, font=font, fill="black")

    # Draw main white text
    draw.text((x, y), cta_text, font=font, fill="white")

    return np.array(img)

# Use ThreadPoolExecutor for parallel processing of frames
with ThreadPoolExecutor() as executor:
    modified_frames = list(executor.map(add_cta_to_frame, frames))

# Save the new animated GIF
pingpong_frames = modified_frames + modified_frames[::-1]
imageio.mimsave(output_path, modified_frames, fps=12, loop=0)

# Display the saved GIF
display(IPImage(filename=output_path))
```

```python
from typing import Dict, Any
from langgraph.graph import StateGraph

def process_video_node(state: Dict[str, Any]) -> Dict[str, Any]:
    # Read prompt from file
    with open("/content/drive/MyDrive/UserInput", "r") as f:
        prompt = f.read().strip()

    # Hardcoded CTA
    cta_text = "BooM"

    # Configuration
    output_path = "/content/drive/MyDrive/output_with_cta.gif"
    height = width = 256
    num_frames = 12

    # Load model (assuming this is cached/preloaded)
    if "pipe" not in state:
        state["pipe"] = DiffusionPipeline.from_pretrained(
            "cerspense/zeroscope_v2_576w",
            torch_dtype=torch.float16,
            use_safetensors=False
        ).to("cuda")

    # Generate frames
    output = state["pipe"](prompt, num_frames=num_frames, height=height, width=width)
    frames = np.array(output.frames)
    frames = np.squeeze(frames)
    frames = (frames * 255).clip(0, 255).astype(np.uint8)

    # Load font
    try:
        font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 72)
    except:
        font = ImageFont.load_default()

    # Process frames
    with ThreadPoolExecutor() as executor:
        modified_frames = list(executor.map(
            lambda frame: add_cta_to_frame(frame, cta_text, font),
            frames
        ))

    # Save GIF
    imageio.mimsave(output_path, modified_frames, fps=12, loop=0)

    return {"output_path": output_path}

def add_cta_to_frame(frame: np.ndarray, cta_text: str, font: ImageFont) -> np.ndarray:
    img = Image.fromarray(frame)
    draw = ImageDraw.Draw(img)

    # Text positioning
    bbox = draw.textbbox((0, 0), cta_text, font=font)
    text_width = bbox[2] - bbox[0]
    text_height = bbox[3] - bbox[1]
    x = (img.width - text_width) // 2
    y = img.height - text_height - 40

    # Add text outline
    for dx in range(-2, 3):
        for dy in range(-2, 3):
            draw.text((x + dx, y + dy), cta_text, font=font, fill="black")

    # Main text
    draw.text((x, y), cta_text, font=font, fill="white")

    return np.array(img)

# Initialize the graph
workflow = StateGraph(process_video_node)
workflow.add_node("process_video", process_video_node)
workflow.set_entry_point("process_video")
app = workflow.compile()
```

```python
import torch
from diffusers import DiffusionPipeline
import numpy as np
import imageio
from PIL import Image, ImageDraw, ImageFont
from concurrent.futures import ThreadPoolExecutor
from typing import Dict, Any
from langgraph.graph import StateGraph

# Preload the model globally to avoid reloading on each invocation
PIPE = DiffusionPipeline.from_pretrained(
    "cerspense/zeroscope_v2_576w",
    torch_dtype=torch.float16,
    use_safetensors=False
).to("cuda")

def process_video_node(state: Dict[str, Any]) -> Dict[str, Any]:
    """LangGraph node that processes video generation with fixed CTA"""
    try:
        # Read prompt from file
        with open("/content/drive/MyDrive/UserInput", "r") as f:
            prompt = f.read().strip()

        # Fixed CTA text
        cta_text = "BooM"

        # Configuration
        output_path = "/content/drive/MyDrive/output_with_cta.gif"
        height = width = 256
        num_frames = 12

        # Generate frames
        output = PIPE(prompt, num_frames=num_frames, height=height, width=width)
        frames = np.array(output.frames)
        frames = np.squeeze(frames)
        frames = (frames * 255).clip(0, 255).astype(np.uint8)

        # Load font
        try:
            font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 72)
        except:
            font = ImageFont.load_default()

        # Process frames in parallel
        with ThreadPoolExecutor() as executor:
            modified_frames = list(executor.map(
                lambda frame: add_cta_to_frame(frame, cta_text, font),
                frames
            ))

        # Save GIF
        imageio.mimsave(output_path, modified_frames, fps=12, loop=0)

        return {"status": "success", "output_path": output_path}

    except Exception as e:
        return {"status": "error", "message": str(e)}

def add_cta_to_frame(frame: np.ndarray, cta_text: str, font: ImageFont) -> np.ndarray:
    """Helper function to add CTA to a single frame"""
    img = Image.fromarray(frame)
    draw = ImageDraw.Draw(img)

    # Text positioning
    bbox = draw.textbbox((0, 0), cta_text, font=font)
    text_width = bbox[2] - bbox[0]
    text_height = bbox[3] - bbox[1]
    x = (img.width - text_width) // 2
    y = img.height - text_height - 40

    # Add text outline
    for dx in range(-2, 3):
        for dy in range(-2, 3):
            draw.text((x + dx, y + dy), cta_text, font=font, fill="black")

    # Main text
    draw.text((x, y), cta_text, font=font, fill="white")

    return np.array(img)

# Initialize and compile the graph
workflow = StateGraph(process_video_node)
workflow.add_node("process_video", process_video_node)
workflow.set_entry_point("process_video")
app = workflow.compile()
```

```python
# Execute the graph (this will process the video)
result = app.invoke({})
print(result)
```
